```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("whitegrid")
```

Graph for $|sinx|< |x|$

```python
x_values = np.linspace(-4,4,1000) 
squares = [abs(np.sin(x)) for x in x_values] 
```

```python

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x_values, squares)
ax.plot(x_values,abs(x_values))


ax.grid(True, which='both')

ax.axhline(y=0, color='k')
ax.axvline(x=np.pi, color='gray')
ax.axvline(x=-np.pi, color='gray')
ax.axvline(x=0, color='k')

ax.annotate("$\pi$",xy=(np.pi, 0),xytext=(np.pi, -0.4),arrowprops=dict(arrowstyle="->",connectionstyle="arc3"))
ax.annotate("$-\pi$",xy=(-np.pi, 0),xytext=(-np.pi - 0.2, -0.4),arrowprops=dict(arrowstyle="->",connectionstyle="arc3"))

# ax.xaxis.set(ticks=np.linspace(-4,4,8))

sns.despine(ax=ax, offset=0)
#plt.savefig('c:/users/aman/desktop/quantum_computing/thesis_content/figures/sinx_less_than_x.png')
```

### Graph for $|sinx|>|\frac{x}{\pi}|$

```python
x_values = np.linspace(-4,4,1000) 
y1_values = [abs(np.sin(x/2)) for x in x_values]
y2_values = [abs(x/np.pi) for x in x_values]
```

```python

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x_values, y1_values)
ax.plot(x_values, y2_values)

ax.fill_between(x_values, y1_values, y2_values, facecolor="orange", alpha=0.2) 


ax.grid(True, which='both')

ax.axhline(y=0, color='k')
ax.axvline(x=np.pi, color='gray')
ax.axvline(x=-np.pi, color='gray')
ax.axvline(x=0, color='k')

ax.annotate("$\pi$",xy=(np.pi, 0),xytext=(np.pi-0.05, -0.1),arrowprops=dict(arrowstyle="->",connectionstyle="arc3"))
ax.annotate("$-\pi$",xy=(-np.pi, 0),xytext=(-np.pi - 0.2, -0.1),arrowprops=dict(arrowstyle="->",connectionstyle="arc3"))

# ax.xaxis.set(ticks=np.linspace(-4,4,8))

sns.despine(ax=ax, offset=0)
#plt.savefig('c:/users/aman/desktop/quantum_computing/thesis_content/figures/sinx__2_less_than_x__pi.png')
```

### Graph for $|sinx|>|k\frac{x}{\pi}|$

```python
x_values = np.linspace(-5,5,1000) 
y1_values = [abs(np.sin(x/2)) for x in x_values]
k = 2/3*np.sin(3*np.pi/4)
y2_values = [abs(x*k/np.pi) for x in x_values]
```

```python

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x_values, y1_values)
ax.plot(x_values, y2_values)

ax.fill_between(x_values, y1_values, y2_values, facecolor="orange", alpha=0.2) 


ax.grid(True, which='both')

ax.axhline(y=0, color='k')
ax.axvline(x=3*np.pi/2, color='gray')
ax.axvline(x=-3*np.pi/2, color='gray')
ax.axvline(x=0, color='k')

ax.annotate("$3\pi/2$",xy=(3*np.pi/2, 0),xytext=(3*np.pi/2-0.05, -0.1),arrowprops=dict(arrowstyle="->",connectionstyle="arc3"))
ax.annotate("$-3\pi/2$",xy=(-3*np.pi/2, 0),xytext=(-3*np.pi/2 - 0.2, -0.1),arrowprops=dict(arrowstyle="->",connectionstyle="arc3"))

# ax.xaxis.set(ticks=np.linspace(-4,4,8))

sns.despine(ax=ax, offset=0)
plt.savefig('c:/users/aman/desktop/quantum_computing/thesis_content/figures/sinx__2_less_than_k_x__pi.png')
```
